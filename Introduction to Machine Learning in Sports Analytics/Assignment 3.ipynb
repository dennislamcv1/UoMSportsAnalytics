### Train Test Split

```python
df.shape
```

```python
training_df=df[0:800]
testing_df=df[800:]
```

```python
X_train=training_df[training_df.columns.drop('outcome_categorical') ]

y_train=training_df['outcome_categorical']

X_test=testing_df[testing_df.columns.drop('outcome_categorical') ]

y_test=testing_df['outcome_categorical']
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

### Using GridSearchCV

```python
model = DecisionTreeClassifier(random_state=1337)
```

```python
parameters = {'max_depth': (3,4,5,6,7,8,9,10),
              'min_samples_leaf': (1,5,10,15,20,25),
             }
```

```python
grids = GridSearchCV(estimator=model,param_grid=parameters,scoring='accuracy',
                     n_jobs=-1,cv=10,verbose=1,return_train_score=True)
```

```python
grids.fit(X_train,y_train)
```

```python
grids.best_params_
```

```python
grids.best_estimator_
```

```python
grids.best_estimator_.score(X_test,y_test)
```
