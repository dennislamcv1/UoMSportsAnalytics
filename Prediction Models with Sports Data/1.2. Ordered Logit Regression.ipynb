### 2. Ordered Logistic Regression

#### To run ordered logit regression in Python, we need to install a new library, "bevel" which provide codes to fit the ordered logit model and obtain the thresholds of each categorical outcome and the coefficient for the IV. 

### 2.1) Import libraries and fit the ordered logit model

```python
#Import bevel library to fit ordit model
from bevel.linear_ordinal_regression import OrderedLogit
import pandas as pd
import numpy as np
```

```python
# Fit the ordered logit model and print the result
ol = OrderedLogit()
ol.fit(NHL_reg_2016['pyth_wpct'], NHL_reg_2016['win.ord'])
ol.print_summary()
```

### 2.2 ) Obtain the coefficients, the intercepts, and the standard errors
The intercepts define the thresholds between a loss and a draw (L|D) and draw and a win (D|W). The probabilities are dependent on (a) the coefficient (i.e., pyth_wpct) and (b) the intercepts. 

```python
# Obtain the coefficients and the intercepts
print(f'beta = {ol.coef_[0]:.4f}')
print(f'interceptL|D = {ol.coef_[1]:.4f}')
print(f'interceptD|W = {ol.coef_[2]:.4f}')
print(ol.coef_)
```

```python
# Standard errors
print(ol.se_)
```

### 2.3) Obtain the fitted values from the model
We can obtain 1) linear product of the ordered regression model, 2) fitted proabilities of being in each outcome of the game and 3) fitted game results from the probabilities. 

```python
# Obtain the linear product from regression
ol.predict_linear_product(NHL_reg_2016[['pyth_wpct']])
```

```python
# Obtain the fitted probabilities of each event 
ol.predict_probabilities(NHL_reg_2016[['pyth_wpct']])
```

```python
# Returns the output with highest probability 
ol.predict_class(NHL_reg_2016[['pyth_wpct']])
```

```python
#Convert array into dataframe to compare with the true values
fitted_results = ol.predict_class(NHL_reg_2016[['pyth_wpct']])
fitted_results_df = pd.DataFrame(fitted_results, columns = ['fitted_results'])
fitted_results_df.shape
```

```python
# Append the fitted results to NHL_reg_2016 data frame to compare them with true values in win.ord column
NHL_reg_2016['fitted_results'] = fitted_results_df

NHL_reg_2016.head()
```

```python
NHL_reg_2016['TRUE'] = np.where(NHL_reg_2016['fitted_results']==NHL_reg_2016['win.ord'], 1, 0)
display(NHL_reg_2016)
```

```python
# Obtain the success rate
Total = NHL_reg_2016['TRUE'].sum()
print(Total/2424)
```

### 2.4) Obtain the fitted values by applying the ordered logit model directly
We obtained 1) linear product of the regression model, 2) fitted proabilities of each event, and 3) fitted game results from the probabilities using python commands above. Alternatively, we can also get all of these statistics by using the regression equation directly. Below you can find that the resulting columns are equavalient to the fitted values that we obtained above. 

```python
# Obtain the fitted probabilities by using the regression coefficients
NHL_reg_2016['predL'] = 1/(1+np.exp(-(ol.coef_[1]-ol.coef_[0]*NHL_reg_2016['pyth_wpct'])))
NHL_reg_2016['predD'] = 1/(1+np.exp(-(ol.coef_[2]-ol.coef_[0]*NHL_reg_2016['pyth_wpct']))) - NHL_reg_2016['predL']
NHL_reg_2016['predW'] = 1- NHL_reg_2016['predL'] - NHL_reg_2016['predD']

# Display the outcome
display(NHL_reg_2016)
```

```python
# Obtain the fitted game results by using the regression coefficients
NHL_reg_2016.loc[(NHL_reg_2016.predL > NHL_reg_2016.predW) & (NHL_reg_2016.predL > NHL_reg_2016.predD), 'fitted_result2'] = 0
NHL_reg_2016.loc[(NHL_reg_2016.predW > NHL_reg_2016.predL) & (NHL_reg_2016.predW > NHL_reg_2016.predD), 'fitted_result2'] = 2
NHL_reg_2016.loc[(NHL_reg_2016.predD > NHL_reg_2016.predW) & (NHL_reg_2016.predD > NHL_reg_2016.predL), 'fitted_result2'] = 1

pd.set_option('display.max_columns',30)

# Display the outcome
display(NHL_reg_2016)
```

### 2.5) Now we assess the accuracy of the ordit model in comparison to the true values

```python
# Obtain the number of fitted values same as the actual outcomes
NHL_reg_2016['TRUE'] = np.where(NHL_reg_2016['fitted_result2']==NHL_reg_2016['win.ord'], 1, 0)
display(NHL_reg_2016)
```

```python
# Obtain the success rate
Total = NHL_reg_2016['TRUE'].sum()
print(Total/2424)
```

### 3) We can also incorporate the home team advantage into the ordered logit model above

```python
iv = ['pyth_wpct', 'home']
```

```python
ol = OrderedLogit()
ol.fit(NHL_reg_2016[iv], NHL_reg_2016['win.ord'])
ol.print_summary()
```

```python
#%% To get the coefficients and the intercepts
print(f'pyth_wpct = {ol.coef_[0]:.4f}')
print(f'home = {ol.coef_[1]:.4f}')
print(f'interceptL|D = {ol.coef_[2]:.4f}')
print(f'interceptD|W = {ol.coef_[3]:.4f}')
print(ol.coef_)
```

```python
NHL_reg_2016['predL2'] = 1/(1+np.exp(-(ol.coef_[2]-ol.coef_[0]*NHL_reg_2016['pyth_wpct']-ol.coef_[1]*NHL_reg_2016['home'])))
NHL_reg_2016['predD2'] = 1/(1+np.exp(-(ol.coef_[3]-ol.coef_[0]*NHL_reg_2016['pyth_wpct']-ol.coef_[1]*NHL_reg_2016['home']))) - NHL_reg_2016['predL']
NHL_reg_2016['predW2'] = 1- NHL_reg_2016['predL'] - NHL_reg_2016['predD']
display(NHL_reg_2016)
```

```python
NHL_reg_2016.loc[(NHL_reg_2016.predL2 > NHL_reg_2016.predW2) & (NHL_reg_2016.predL2 > NHL_reg_2016.predD2), 'fitted'] = 0
NHL_reg_2016.loc[(NHL_reg_2016.predW2 > NHL_reg_2016.predL2) & (NHL_reg_2016.predW2 > NHL_reg_2016.predD2), 'fitted'] = 2
NHL_reg_2016.loc[(NHL_reg_2016.predD2 > NHL_reg_2016.predW2) & (NHL_reg_2016.predD2 > NHL_reg_2016.predL2), 'fitted'] = 1
```

```python
NHL_reg_2016['TRUE2'] = np.where(NHL_reg_2016['fitted']==NHL_reg_2016['win.ord'], 1, 0)
display(NHL_reg_2016)
```

```python
Total = NHL_reg_2016['TRUE2'].sum()
print(Total/2424)
```
