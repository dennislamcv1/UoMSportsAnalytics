# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("Game logs 1999-2003.xlsx")
```

```python
# Dummy variables for home and away wins
Teams['hwin'] = np.where(Teams['home_score'] > Teams['visitor_score'],1,0)
Teams['awin'] = np.where(Teams['home_score'] < Teams['visitor_score'],1,0)
```

```python
# Extract year from date
Teams['year'] = Teams['date'].astype(str).str[0:4]
display(Teams)
```

```python
# Aggregate stats for home team
Teamshome = Teams.groupby(['home','year'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 'visitor_hr', 'visitor_sf',
                                           'visitor_hbp','visitor_bb','home_ab', 'home_h', 'home_2b', 'home_3b',
                                           'home_hr', 'home_sf', 'home_hbp', 'home_bb', 'hwin','awin'].sum()
Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"home": "Team"}) 
Teamshome
```

```python
# Aggregate stats for away team
Teamsaway = Teams.groupby(['year','visitor'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 
                                           'visitor_hr', 'visitor_sf', 'visitor_hbp', 'visitor_bb', 
                                           'home_ab', 'home_h', 'home_2b', 'home_3b', 'home_hr', 
                                           'home_sf', 'home_hbp', 'home_bb','hwin','awin'].sum()

Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor": "Team"}) 
Teamsaway
```

```python
# Merge into one dataframe
Teams2 = pd.merge(Teamshome, Teamsaway, on =['Team','year'])
pd.set_option('display.max_columns', 50)
display(Teams2)
```

```python
# Total Wins
Teams2['wins'] = Teams2['hwin_x']+Teams2['awin_y']
```

```python
MedWins = Teams2.groupby(['year'])['wins'].median()
MedWins
```

```python
# Home hits
HomeHits = Teamshome.groupby(['year'])['home_h'].min()
HomeHits
```

```python
# Avg AB
AwayAB = Teamsaway.groupby(['year'])['visitor_ab'].mean()
AwayAB
```

# II. Calculating Performance Metrics

```python
# Create batting average for and batting average against
Teams2['AVGFOR'] = (Teams2['home_h_x'] + Teams2['visitor_h_y'])/(Teams2['home_ab_x'] + Teams2['visitor_ab_y'])
Teams2['AVGAGN'] = (Teams2['home_h_y'] + Teams2['visitor_h_x'])/(Teams2['home_ab_y'] + Teams2['visitor_ab_x'])
```

```python
# Calculate win percentage
Teams3 = Teams2[['year', 'Team', 'wins','AVGFOR','AVGAGN']]

TeamsGh = Teams.groupby(['year','home'])['hwin'].count()
TeamsGh = TeamsGh.reset_index()
TeamsGh = TeamsGh.rename(columns = {"home": 'Team', "hwin": 'hgame'}) 

TeamsGa = Teams.groupby(['year','visitor'])['awin'].count()
TeamsGa = TeamsGa.reset_index()
TeamsGa = TeamsGa.rename(columns = {"visitor": 'Team', "awin": 'agame'}) 

TeamsG = pd.merge(TeamsGh, TeamsGa, on =['Team','year'])
TeamsG['Games'] = TeamsG['hgame']+TeamsG['agame']
TeamsG = TeamsG[['year','Team','Games']]

Teams3 = pd.merge(Teams3, TeamsG, on =['Team','year'])
Teams3['wpc'] = Teams3['wins']/Teams3['Games']
Teams3
```

```python
# Calculate home win percentage and away win percentage
TeamsWh = Teams.groupby(['year','home'])['hwin'].sum()
TeamsWh = TeamsWh.reset_index()
TeamsWh = TeamsWh.rename(columns = {"home": 'Team'}) 
Teamsh = pd.merge(TeamsGh, TeamsWh, on=['Team','year'])
Teamsh['hwpc'] = Teamsh['hwin']/Teamsh['hgame']

TeamsWa = Teams.groupby(['year','visitor'])['awin'].sum()
TeamsWa = TeamsWa.reset_index()
TeamsWa = TeamsWa.rename(columns = {"visitor": 'Team'})
Teamsa = pd.merge(TeamsGa, TeamsWa, on=['Team','year'])
Teamsa['awpc'] = Teamsa['awin']/Teamsa['agame']

```

```python
# Merge home and away win percentage data frames and calculate difference
Teamsha = pd.merge(Teamsh, Teamsa, on=['Team','year'])
Teamsha['wpc_diff'] = Teamsha['hwpc'] - Teamsha['awpc']
Teamsha
```

```python
# Maximum WPCT diff
Teamsha.sort_values(by=['wpc_diff'], ascending = False)
```

```python
# AVG Diff
Teams3['AVGDIFF'] = Teams3['AVGFOR'] - Teams3['AVGAGN']
Teams3.sort_values(by=['AVGDIFF'])
```
