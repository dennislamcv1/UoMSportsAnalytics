# I. Building the Dataset: Player Stats

```python
# First we load the packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in salary data retrieved from Lahman's Database
Salary = pd.read_csv("Salaries.csv")
Salary = Salary[Salary.salary > 0] 
display(Salary)
```

```python
# Calculate natural log of salary 
Salary['lnSal'] = np.log(Salary['salary'])
Salary
```

```python
# Rename 'YearID' as 'SalYear'
Salary = Salary.rename(columns = {'yearID':'SalYear'})
Salary
```

```python
# Let's create a copy of the df which we call "Master", which we can then use to merge the information from all the other dfs
Master = Salary
Master
```

```python
# Read in batting

Batting = pd.read_csv("Batting.csv")
pd.set_option('display.max_columns', 50)
display(Batting)
```

```python
# Using '.groupby' to sum across stints

Batting = Batting.groupby(['playerID','yearID']).sum()
Batting.reset_index(inplace=True)
Batting
```

```python
Batting = Batting[(Batting.yearID >= 1998) & (Batting.yearID <= 2006) & (Batting.AB >= 130)].copy()
display(Batting)
```

```python
# Calculate OBP and SLG in Batting data

Batting['PA'] = Batting['AB'] + Batting['BB'] + Batting['HBP'] + Batting['SH'] + Batting['SF']

Batting['OBP'] = (Batting['H'] + Batting['BB'] + Batting['HBP'])/(Batting['AB'] + Batting['BB']
                                                                  + Batting['HBP'] + Batting['SF'])

Batting['SLG'] = ((Batting['H'] - Batting['2B'] - Batting['3B'] - Batting['HR']) 
                  + 2*Batting['2B'] + 3*Batting['3B'] + 4*Batting['HR'])/Batting['AB']

Batting['AVG'] = Batting['H']/Batting['AB']

display(Batting)
```

```python
# Create another year variable to link previous season stats to next year's salary in batting data

Batting['SalYear'] = Batting['yearID'] + 1 
display(Batting)
```

```python
# carrying out the merge
Master = pd.merge(Batting, Master, on=['SalYear', 'playerID'])
display(Master)
```

```python
# Calculate average player salary by season
MeanSal = Master.groupby(['SalYear'])['salary'].mean().reset_index()
display(MeanSal)
```

```python
# Calculate average player OBP and SLG by season
Mean_OBPSLG = Master.groupby(['yearID'])['OBP','SLG'].mean()
display(Mean_OBPSLG)
max_OBP = max(Mean_OBPSLG['OBP'])
display(max_OBP)
max_SLG = max(Mean_OBPSLG['SLG'])
display(max_SLG)
```

```python
# Calculate aggregate player HR across timeframe
HR_Timeframe = Master.groupby(['playerID'])['HR'].sum().reset_index()
HR_Timeframe= HR_Timeframe.sort_values(by=['HR'],  ascending=False)
display(HR_Timeframe)
#max(HR_Timeframe)
```

# II. Building the Dataset: Player Info

```python
# Read in People data retrieved from Lahman's Database

People = pd.read_csv("People.csv")
Debut = People[['playerID','debut']].copy()
Debut['debutyr'] = Debut['debut'].astype(str).str[0:4]
Debut = Debut[['playerID','debutyr']]

display(Debut)
```

```python
# Now merge debut year into master data
Master = pd.merge(Master, Debut, on=['playerID'], how = 'left')
display(Master)
```

```python
# Use experience as proxy to classify players contract status as arbitration eligible or free agent
Master['Exp'] = Master['yearID'] - Master['debutyr'].astype(int)
Master['Arb'] = np.where((Master['Exp'] <= 6) & (Master['Exp'] >= 3),1,0)
Master['Free'] = np.where(Master['Exp'] > 6, 1, 0)
display(Master)
```

```python
# Read in player appearance data retrieved from Lahman's Database

Appearances = pd.read_csv("Appearances.csv")
Appearances = Appearances.groupby(['playerID','yearID'])['G_c','G_1b','G_2b','G_3b','G_ss','G_of','G_dh'].sum()                                       
Appearances.reset_index(inplace=True)
Appearances['Max_G'] = Appearances[["G_c","G_1b","G_2b","G_3b","G_ss","G_of","G_dh"]].max(axis=1)
display(Appearances)
```

```python
# (ii) Assign each player season to a primary position 

# Create definiton to position classification process
def Position(df):
    if (df['Max_G'] == df['G_c']): return "C"
    elif (df['Max_G'] == df['G_1b']): return "1B"
    elif (df['Max_G'] == df['G_2b']): return "2B"
    elif (df['Max_G'] == df['G_3b']): return "3B"
    elif (df['Max_G'] == df['G_ss']): return "SS"
    elif (df['Max_G'] == df['G_of']): return "OF"
    elif (df['Max_G'] == df['G_dh']): return "DH"

# Apply newly created definition to appearance data as variable 'POS'
Appearances['POS'] = Appearances.apply(Position, axis = 1)    

display(Appearances)
```

```python
# Exclude non-position players and keep necessary variables

Appearances = Appearances[Appearances['Max_G'] > 0] 
Appearances = Appearances[['playerID','yearID','POS']]
display(Appearances)
```

```python
# Create a indicator variable for catcher and infielder (2B, 3B, SS)

Appearances['Catch'] = np.where(Appearances['POS'] == "C", 1, 0)
Appearances['SecondB'] = np.where((Appearances['POS'] == "2B"), 1, 0)
Appearances['ThirdB'] = np.where((Appearances['POS'] == "3B"), 1, 0)
Appearances['SS'] = np.where((Appearances['POS'] == "SS"), 1, 0)

display(Appearances)
```

```python
# Merge position data into Master data

Master = pd.merge(Master, Appearances, on=['playerID','yearID'], how = 'left')
display(Master)
```

```python
# Determine highest paid position in 1999 and 2004
POSSal = Master.groupby(['SalYear','POS'])['salary'].mean().reset_index()
POSSal= POSSal.sort_values(by=['salary'],  ascending=False)
POSSal_9904 = POSSal[(POSSal['SalYear'] == 1999) | (POSSal['SalYear'] == 2004)]
display(POSSal_9904)
```

```python
# Proportion of arbitration eligible or free agent eligible players
Arb_Total = sum(Master['Arb'])
FA_Total = sum(Master['Free'])
Total = len(Master['playerID'])
Prop = (Arb_Total + FA_Total)/Total
Prop
```

```python
# Sum years of experience by team in 2002
Master_2002 = Master[Master['yearID']==2002]
Master_2002_Exp = Master_2002.groupby(['teamID'])['Exp'].sum().reset_index()
Master_2002_Exp = Master_2002_Exp.sort_values(by=['Exp'],ascending=False)
display(Master_2002_Exp)
```

# III. Running Regressions

```python
# Run regression for Pre-MB Era
import statsmodels.formula.api as smf
MB_Data_Pre = Master[(Master.SalYear >= 1999) & (Master.SalYear <= 2003)]
Pre_MB_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_Pre).fit()
Pre_MB_lm.summary()
```

```python
# Run regression for Post-MB Era
MB_Data_Post = Master[(Master.SalYear >= 2004) & (Master.SalYear <= 2006)]
Post_MB_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_Post).fit()
Post_MB_lm.summary()
```

```python
# Regression for all seasons individually
MB_Data_1999 = Master[(Master.SalYear == 1999)]
Val_1999_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_1999).fit()
MB_Data_2000 = Master[(Master.SalYear == 2000)]
Val_2000_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2000).fit()
MB_Data_2001 = Master[(Master.SalYear == 2001)]
Val_2001_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2001).fit()
MB_Data_2002 = Master[(Master.SalYear == 2002)]
Val_2002_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2002).fit()
MB_Data_2003 = Master[(Master.SalYear == 2003)]
Val_2003_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2003).fit()
MB_Data_2004 = Master[(Master.SalYear == 2004)]
Val_2004_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2004).fit()
MB_Data_2005 = Master[(Master.SalYear == 2005)]
Val_2005_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2005).fit()
MB_Data_2006 = Master[(Master.SalYear == 2006)]
Val_2006_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + AVG + PA + Arb + Free + Catch + SecondB + ThirdB + SS', data=MB_Data_2006).fit()
```

```python
# Create Pre-MB table to display results
from statsmodels.iolib.summary2 import summary_col
Header = ['1999','2000','2001','2002','2003','2004','2005','2006']
MB_Table = summary_col([Val_1999_lm,Val_2000_lm,Val_2001_lm,Val_2002_lm,Val_2003_lm,
                            Val_2004_lm,Val_2005_lm,Val_2006_lm,],
                           stars=True,float_format="'%.3f'",model_names = Header)
print(MB_Table)
```
